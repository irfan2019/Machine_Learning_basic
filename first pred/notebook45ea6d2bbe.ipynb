```python
import pandas as pd
import numpy as np
import matplotlib as mlt
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/carmodel/Cardheko.csv")
dff=df.drop(['fuel','sellertype','transmission'],axis='columns')
dff.drop("name",axis="columns")
```

```python
from sklearn.linear_model import LinearRegression
x=dff[["year","owner","kmdriven"]]
y=dff.sellingprice
model=LinearRegression()
model.fit(x,y)
print("y=mx+c")
print("c is ",model.intercept_)
print("m is ",model.coef_)
```

```python
prediction=model.predict([[2007,2,600000]])
print("selling price is",prediction)
```
